## Clean bad residuals

```python
counts_min = 40
grain      = 1
clean = np.copy(sel)
done  = False
while (not done):
    coors  = mkdst[clean].x, mkdst[clean].y
    energy = mkdst[clean].energy
    dtime  = mkdst[clean].dtime
    krmap, residuals  = krana.krmap(coors, dtime, energy, 
                                    bins = (20 * grain, 20 * grain), 
                                    counts_min = counts_min, dt0 = None)
    done, usel = krana.accept_residuals(residuals, nsigma = 3.5, min_sigma = 0.99, plot = True)
    clean[clean == True] = usel
    print('cleanning ', done, np.sum(clean), np.sum(clean)/len(clean))

print('len ', len(clean), len(sel))
sels['clean'] = clean
```

## Kr map with clean residuals

```python
sel    = sels['clean']
coors  = mkdst[sel].x, mkdst[sel].y
energy = mkdst[sel].energy
dtime  = mkdst[sel].dtime
counts_min = 40
grain      = 2
krmap, residuals = krana.krmap(coors, dtime, energy, 
                               bins = (18 * grain, 18 * grain),
                               counts_min = counts_min, dt0 = None)
```
